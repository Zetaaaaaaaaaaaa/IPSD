```python
#Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_csv("C:/Users/acer/Downloads/archive/Ecommerce_Sales_Data_2024_2025.csv")

# 2. Basic Information About Dataset
df.info()

# 3. Cek Nilai Duplikat, Nilai Unik
df.drop_duplicates(inplace=True)

for i in df.columns:
  print(f"Kolom : {i}")
  print(f"Jumlah : {df[i].nunique()}")
  print(f"Unik : {df[i].unique()}")
  
# 4. Visualisasikan Jumlah Nilai Unik
Total_Unique = df.nunique()

plt.figure(figsize=(16, 9))
Total_Unique.plot(kind='bar', color='#3498db', edgecolor='black')
plt.title('Jumlah Nilai Unik Setiap Kolom', fontsize=16, weight='bold', color='#2c3e50')
plt.xlabel('Kolom', fontsize=12)
plt.ylabel('Jumlah Nilai Unik', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# 5. Menemukan Null Values

df.isnull().sum()
```

```python
# 6. Replace Semua Null Values

import pandas as pd

print("Jumlah nilai null sebelum diganti:\n", df.isnull().sum())

df = df.fillna({
    'Customer Name': 'Tidak diketahui',
    'City': 'Tidak diketahui',
    'Profit': 0,
})

print("\nJumlah nilai null setelah diganti:\n", df.isnull().sum())

df.to_csv("C:/Users/acer/Downloads/archive/Ecommerce_Sales_Data_2024_2025.csv", index=False)
```

```python
# 7. Melihat Tipe Dataset

df.dtypes
```

```python
# 8. Filter Data

dFiltered = df[df['Order ID'] > 10000]
dFiltered.head()
```

```python
# 9. Boxplot Kolom Numerik

numerik_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(16, 9))
sns.set(style="whitegrid")
sns.boxplot(data=numerik_df, palette="viridis", linewidth=1.2)
plt.title("Distribusi Kolom Numerik", fontsize=14)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 10. Visualisasi korelasi

train_df = df.copy()
corr_matrix = train_df.select_dtypes(include='number').corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.tight_layout()
plt.show()
```
